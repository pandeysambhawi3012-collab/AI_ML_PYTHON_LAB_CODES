```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

n = 100

df = pd.DataFrame({
    "Years_of_Experience": np.random.randint(1, 15, n),
    "Salary": np.random.randint(20000, 100000, n)
})

df.to_csv("salary_dataset11.csv", index=False)
df = pd.read_csv("salary_dataset11.csv")
```

```python
X = df[["Years_of_Experience"]]
y = df["Salary"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
print("\nMAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
plt.scatter(X_test, y_test, label="Actual Data")
plt.plot(X_test, y_pred, color='red', label="Regression Line")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression")

plt.legend()
plt.show()
```
